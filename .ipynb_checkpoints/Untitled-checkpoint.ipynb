```python
import pandas as pd
```

```python
df= pd.read_csv("preprocessed_data.csv")
```

```python
df= df.dropna()
df['topics']= df['topics'].str.split(',').str[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#define vectorizer parameters
tfidfconverter = TfidfVectorizer(max_features=1500, min_df=5, max_df=0.7)
tfidf_matrix = tfidfconverter.fit_transform(df["processed_text"]) #fit the vectorizer to synopses

print(tfidf_matrix.shape)
```

```python
terms = tfidfconverter.get_feature_names()
```

```python
tfidf_matrix=tfidf_matrix.toarray()
```

```python
from sklearn.cluster import KMeans

num_clusters = 5

km = KMeans(n_clusters=num_clusters)

km.fit(tfidf_matrix)

clusters = km.labels_.tolist()
```

```python
clusters
```

```python
df['clusters']=clusters
df['clusters'].value_counts()
```

```python
grouped = df.groupby(df['clusters'])
```

```python
print(grouped.mean())
```

```python
for i in grouped:
    print(i)
```

```python
datasets = {}
for groups, data in grouped:
        datasets[groups] = data
```

```python
from sklearn.externals import joblib
X=[]
def cluster(data, num, file):
    matrix= tfidfconverter.fit_transform(data["processed_text"])
    X.append(matrix)
    joblib.dump(matrix,'C:/Users/raksh/Desktop/assignment/bigdata project/models/tfidf'+file+'.pkl')
    km = KMeans(n_clusters=num)
    km.fit(matrix)
    joblib.dump(km,'C:/Users/raksh/Desktop/assignment/bigdata project/models/model'+file+'.pkl')
    clusters = km.labels_.tolist()
    data['new_cluster']=clusters
    data.to_csv("C:/Users/raksh/Desktop/assignment/bigdata project/new_datasets/"+str(file)+".csv")
    return data
```

```python
#cluster(datasets[0],len(datasets[0]['topics'].unique()), str(1))
for i in range(0,5):
    cluster(datasets[i], len(datasets[i]['topics'].unique()), str(i))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracys=[]
def models(i):
    X_train, X_test, y_train, y_test = train_test_split(X[i], datasets[i]['topics'], test_size=0.33, random_state=42)
    clf= MultinomialNB().fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    accuracy={
        "accuracy": round(accuracy_score(y_test, y_pred)*100,2),
        "pscore": round(precision_score(y_test, y_pred, average='macro')*100,2),
        "rscore":round(recall_score(y_test, y_pred, average='macro')*100,2),
        "f1score":round(f1_score(y_test, y_pred, average='macro')*100,2)
        }
    accuracys.append(accuracy)
```
